```python
# importing necessary packages 
import pandas as pd 
from apyori import apriori 
```

```python
# Read data and Display 
store_data = pd.read_csv('store_data.csv', header=None) 
 
print("Rows and columns of the data :: ",store_data.shape) 
display(store_data.head())
```

```python
# Preprocessing on Data 
#Here we need a data in form of list for Apriori Algorithm. 
records = [] 
for i in range(0, 7501): 
    records.append([str(store_data.values[i, j]) for j in range(0, 20) if str(store_data.values[i, j])!= 'nan'])
```

```python
# for Apriori Algorithm 
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2,min_lift=3, min_length=2) 
association_results = list(association_rules) 
# printing number of relations derived 
print("There are {} Relation derived.".format(len(association_results))) 
```

```python
# Post procesing of Association Rules Derived 
# Printing Rules along with Support, Confidence and Lift
rno=1
for i in range(0, len(association_results)):
    r=list(association_results[i][0])[0] +" ---> "+list(association_results[i][0])[1]
    s=association_results[i][1]
    c=str(association_results[1][2][0]).split(",")[-2].split("=")[1]
    l=str(association_results[i][2][0]).split(",")[-1].split("=")[1].split(")")[0]
    print("\n======================================== RULE - ",rno, " ================================================")
    print("\nRULE          :: ", r,"\n")
    print("SUPPORT       :: ", s)
    print("CONFIDENCE    :: ", c)
    print("LIFT          :: ", l)
    rno+=1      
```
